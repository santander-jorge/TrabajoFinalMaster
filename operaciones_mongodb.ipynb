```python
import pandas as pd
from pymongo import MongoClient
from pymongo.server_api import ServerApi

# Funci√≥n para limpiar los datos
def limpiar_datos(data):
    # Convertir NaN a None
    for doc in data:
        for key, value in doc.items():
            if pd.isna(value):
                doc[key] = None
            # Remover citas dobles al inicio y al final de las cadenas
            elif isinstance(value, str) and value.startswith('"') and value.endswith('"'):
                doc[key] = value[1:-1]
            # Convertir valores num√©ricos en cadenas a n√∫meros
            elif isinstance(value, str) and value.replace('.', '', 1).isdigit():
                doc[key] = float(value) if '.' in value else int(value)
    return data

uri = "mongodb+srv://Jorge:abcdefg@cursoue.vcazaim.mongodb.net/?retryWrites=true&w=majority&appName=CursoUE"

client = MongoClient(uri, server_api=ServerApi('1'))
db = client['RNA_database']  # Reemplaza 'mi_base_de_datos' con el nombre de tu base de datos
collection = db['RNA_coleccion']  # Reemplaza 'mi_coleccion' con el nombre de tu colecci√≥n


# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    raise Exception("Unable to connect to the deployment. Check your connection string. Error:", e)

# Leer el CSV
df = pd.read_csv(r"C:\Users\Usuario\Desktop\final\finales\Final2.csv", sep=';', low_memory=False)  
print(len(df))
# Convertir el DataFrame a una lista de diccionarios
data = df.to_dict(orient='records')
data = limpiar_datos(data)
print(len(data))
```
